```python
import glob
import pandas as pd
import io

# absolute path to search all text files inside a specific folder
path = r'D:/temp/jalan*.shp'
jalan=[]
files = glob.glob(path)

for file in files:
    jalan.append(files)
    
#print(files)
# len(files)
jalan_df = pd.DataFrame(jalan[0])
# j_df=pd.concat([jalan_df.T[x] for x in jalan_df.T], ignore_index=True)
# jalan_df2 = j_df.to_frame()

jalan_df2=jalan_df

jalan_df2.columns[0]
jalan_df2.columns = ['path']
jalan_df2

jalan_df = pd.DataFrame(jalan_df2, columns= ['path'])
jalan_df2['filename'] = jalan_df['path'].str[8:80]
jalan_df2['grid'] = jalan_df2['path'].str[14:80]
jalan_df2 = jalan_df2.astype({'path':'string','filename':'string','grid':'string'})

jalan_df2.head()
jalan_df2
```
